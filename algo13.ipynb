### An example of a greedy attribute ranking algorithm. Examples of the MI function are Definition 4.18 or Algorithm 10

```python
import pandas as pd
from math import log2
```

```python
# Load the Titanic dataset
data = pd.read_csv('titanic_train.csv')
```

```python
data
```

```python
def entropy(y):
    """
    Calculate the entropy of the target variable.
    """
    counts = y.value_counts()
    probs = counts / len(y)
    entropy = sum(-p * log2(p) for p in probs)
    return entropy

def information_gain(X, y, feature):
    """
    Calculate the information gain of a feature.
    """
    total_entropy = entropy(y)
    feature_entropy = 0
    for val in X[feature].unique():
        subset = y[X[feature] == val]
        prob = len(subset) / len(y)
        feature_entropy += prob * entropy(subset)
    return total_entropy - feature_entropy

def attribute_rank(data):
    """
    Implement the greedy attribute ranking algorithm.
    """
    features = data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1).columns
    X = data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
    y = data['Survived']
    ranking = []
    while len(ranking) < len(features):
        remaining = [f for f in features if f not in ranking]
        gain_values = [information_gain(X, y, f) for f in remaining]
        best_feature = remaining[gain_values.index(max(gain_values))]
        ranking.append(best_feature)
    return ranking
```

```python
# Run the attribute ranking algorithm
ranked_features = attribute_rank(data)
```

```python
# Print ranked features with a title-like formatting
print("Parameters ranked by influence on f(x)")
print("=" * 40)  # Line to visually separate title from list
for i, feature in enumerate(ranked_features, start=1):
    print(f"{i}. {feature}")
```
